```python
# Split the x1 string by newline characters, strip whitespace, and filter out empty strings
values = [v for v in x1.strip().split('\n') if v.strip()]

# Ensure we have enough elements
if len(values) < 4:
    raise ValueError("x1 string does not contain enough forecast data.")

# Extract the forecasts
next_year_sales_growth = float(values[0].strip().rstrip('%'))  # Next year's sales growth
next_year_opm = float(values[1].strip().rstrip('%'))  # Next year's operating profit margin
average_5yr_sales_growth = float(values[2].strip().rstrip('%'))  # 5-year average sales growth
average_5yr_opm = float(values[3].strip().rstrip('%'))  # 5-year average operating profit margin

# Years for forecast
forecast_years = ['2025-03-31', '2026-03-31', '2027-03-31', '2028-03-31', '2029-03-31']

# Step 1: Forecast Total Revenue for the next 5 years
# Find the latest date (column) in the DataFrame
latest_date = financial_summary1.columns[0]

# Select the latest "Total Revenue" value dynamically
last_total_revenue = financial_summary1.loc['Total Revenue', latest_date]

forecasted_revenue = [last_total_revenue * (1 + (next_year_sales_growth / 100))]  # First year based on next year's growth

# Apply the average growth rate for the remaining years
for _ in range(1, len(forecast_years)):
    previous_sales = forecasted_revenue[-1]
    next_sales = previous_sales * (1 + average_5yr_sales_growth / 100)  # Use average 5-year sales growth
    forecasted_revenue.append(next_sales)

# Step 2: Forecast EBIT Margin and EBIT
forecasted_ebit_margin = [next_year_opm / 100]  # First year based on next year's OPM
forecasted_ebit_margin += [average_5yr_opm / 100] * (len(forecast_years) - 1)  # Apply average OPM for remaining years

forecasted_ebit = [revenue * margin for revenue, margin in zip(forecasted_revenue, forecasted_ebit_margin)]

# Step 3: Calculate Depreciation as a % of Revenue
avg_depreciation_pct = np.mean(financial_summary1.loc["Depreciation as % of Revenue"])
forecasted_depreciation_pct = [avg_depreciation_pct] * len(forecast_years)

# Step 4: Calculate Tax Rate
avg_tax_rate = np.mean(financial_summary1.loc["Tax Rate"])
forecasted_tax_rate = [avg_tax_rate] * len(forecast_years)

# Step 5: Calculate Investment Rate, filtering out positive outliers
investment_rates = financial_summary1.loc["Investment Rate"]
mean_investment_rate = np.mean(investment_rates)
filtered_investment_rates = investment_rates[
    (investment_rates <= 0) | (investment_rates <= 3 * mean_investment_rate)
]
avg_investment_rate = np.mean(filtered_investment_rates)
forecasted_investment_rate = [avg_investment_rate] * len(forecast_years)

# Step 6: Calculate Free Cash Flow using the formula
forecasted_free_cash_flow = [
    ebit * (1 - tax_rate) + (ebit * investment_rate) + (revenue * dep_pct)
    for ebit, tax_rate, investment_rate, revenue, dep_pct in zip(
        forecasted_ebit, forecasted_tax_rate, forecasted_investment_rate, forecasted_revenue, forecasted_depreciation_pct)
]

# Step 7: Create the forecast DataFrame
forecast_df = pd.DataFrame({
    'Total Revenue': forecasted_revenue,
    'EBIT': forecasted_ebit,
    'Depreciation as % of Revenue': forecasted_depreciation_pct,
    'EBIT Margin': forecasted_ebit_margin,
    'Tax Rate': forecasted_tax_rate,
    'Investment Rate': forecasted_investment_rate,
    'Free Cash Flow': forecasted_free_cash_flow
}, index=forecast_years)

# Combine the forecasted data with the original financial summary
financial_summary1_forecasted = pd.concat([financial_summary1, forecast_df.T], axis=1)

# Display the final DataFrame
financial_summary1_forecasted
```
