```python
import streamlit as st
import nbformat

# Set the title of your app
st.title("Your Investment Friend")

# Path to your Jupyter Notebook
notebook_path = r"C:\Users\Asus\python\RELATIVE VALUATION-Copy1.ipynb"

# Load the notebook
with open(notebook_path, 'r', encoding='utf-8') as f:
    notebook = nbformat.read(f, as_version=4)

# Extract code cells
code_cells = [cell for cell in notebook.cells if cell.cell_type == 'code']

# Execute all code cells
for i, cell in enumerate(code_cells, start=1):  # start=1 to match your selected_cells indices
    exec(cell.source)

    # Try to extract and display the output
    try:
        if cell.outputs:
            output = cell.outputs[0]
            if 'text' in output:
                st.write(f"**Output of cell {i}:**")
                st.text(output['text'])
            elif 'data' in output and 'text/plain' in output['data']:
                st.write(f"**Output of cell {i}:**")
                st.text(output['data']['text/plain'])
            else:
                st.write(f"**Cell {i} executed but no textual output was found.**")
        else:
            st.write(f"**Cell {i} executed with no output.**")
    except Exception as e:
        st.write(f"**Error in displaying output for cell {i}: {e}**")
```
